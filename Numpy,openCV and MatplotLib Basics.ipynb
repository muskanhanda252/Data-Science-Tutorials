# Basics of Numpy

```python
import numpy as np
```

```python
a=np.array([1,2,3,2,4,4,3,4,2,2,2,3,2,4,2,4,5,3,4,5])
a
```

```python
b=np.unique(a,return_counts=True)
b
```

# Now find out the index of the element with the maximum frequency.

```python
c=np.array(b[1])
```

```python
c
```

```python
type(c)
```

```python
idx=np.argmax(c)
print(idx)
```

```python
element_with_highest_frequency=b[0][idx]
element_with_highest_frequency
```

```python
import matplotlib.pyplot as plt
```

```python
x=np.array([1,2,3,4,5,6,7,8])
y=x**2+5
print(x,y)
```

# Different types of styles available in Pyplot:

```python
plt.style.available   
```

```python
plt.style.use('dark_background')
plt.plot(x,y,label="See the line with the help of this")
plt.scatter(x,x**2,label="This is the square line")
plt.title("Matplotlib uses")
plt.xlabel("Numbers")
plt.ylabel("Numbers and their squares + 5")
plt.legend()   #if we don't use legend then the label of the line will not be visible at all.
plt.show()
```

## How to Crop the image in OpenCV?

```python
cropping=rgb_format[100:500,150:450,:] #remember to use the square brackets, also row, column, channel
plt.imshow(cropping)
plt.axis("off")
plt.show()
```

```python
cropping.shape
```

# Resizing the image:

```python
cropped_image=cv2.resize(cropping,(100,100))
```

```python
plt.imshow(cropped_image)
plt.axis("off")
plt.title("A useless watch"+str(cropped_image.shape))
plt.show()
```
